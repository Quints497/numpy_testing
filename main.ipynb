```python
from io import BytesIO

import cv2
import numpy as np

image = cv2.imread('dog.jpeg', cv2.IMREAD_COLOR)

# Save numpy array to a BytesIO buffer
buffer = BytesIO()
np.save(buffer, image)

# Get string representation
x_str = buffer.getvalue()

# Convert string back to numpy array
buffer = BytesIO(x_str)
x_array = np.load(buffer)

print(x_array)
print(x_array.shape)
print(x_array.dtype)
print(type(x_array))


new_image = cv2.cvtColor(x_array, cv2.COLOR_BGR2RGB)
```
